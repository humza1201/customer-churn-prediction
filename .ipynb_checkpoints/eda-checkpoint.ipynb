```python
import matplotlib.pyplot as plt
```

```python

y = df['tenure']
bins = 30
```

```python
plt.hist(y , bins=bins , edgecolor = 'black', color='skyblue', label='tenure distributiion')
plt.xlabel("tenure  (months)")
plt.ylabel("no of customer")
plt.title("performing EDA plotting histogram of tenure !")
plt.legend()
plt.tight_layout()
plt.grid(True)
```

```python
plt.hist(df[df['Churn']=='No']['tenure'], bins=30, alpha=0.7, label='No Churn', color='skyblue', edgecolor='black')
plt.hist(df[df['Churn']=='Yes']['tenure'], bins=30, alpha=0.7, label='Churn', color='salmon', edgecolor='black')

plt.xlabel("Tenure (months)")
plt.ylabel("Number of Customers")
plt.title("Tenure Distribution by Churn Status")
plt.legend()
plt.tight_layout()
plt.grid(True)
```

```python
ten_bins = [0,12, 24, 48, 72]
```

```python
df['tenure_bin'] = pd.cut(df['tenure'], bins=ten_bins, labels=["0–12", "12–24", "24–48", "48–72"], right=False)
```

```python
churn_rate_per_bin = df.groupby('tenure_bin',observed=True)['Churn'].apply(lambda x: (x == 'Yes').mean() * 100)
print(churn_rate_per_bin)
```

```python
bins = [0, 35, 70, 120]
df['monthly_bins'] = pd.cut(df['MonthlyCharges'] , bins = bins, labels=["0–35", "35–70", "70–120"], right=False)
```

```python
churn_rate_by_monthlybins
```

```python
numerical_features = df.select_dtypes(include=['int64','float64']).columns.tolist()
binary_features = [col for col in df.columns if df[col].nunique() == 2]
multi_cat_features = [col for col in df.columns if df[col].dtype == 'object' and df[col].nunique() > 2]

print("Numerical Features:", len(numerical_features)-1) #excluding seniorcitizen
print("Binary Categorical Features:", len(binary_features)-1) #excluding churn
print("Multi-Category Categorical Features:", len(multi_cat_features)-1) # excluding customer id
```
